```python
# Clean & Prepare
labs48_df['charttime'] = pd.to_datetime(labs48_df['charttime'])
labs48_df.dropna(subset=['valuenum'], inplace=True)

# ðŸ’¡ Tip: restrict to useful features before pivoting
labs48_topitems = (
    labs48_df.groupby('label')
    .size()
    .sort_values(ascending=False)
    .head(30)
    .index.tolist()
)
labs48_topitems

# Clean & Prepare
vitals48_df['charttime'] = pd.to_datetime(vitals48_df['charttime'])
vitals48_df.dropna(subset=['valuenum'], inplace=True)

# ðŸ’¡ Tip: restrict to useful features before pivoting
vitals48_topitems = (
    vitals48_df.groupby('label')
    .size()
    .sort_values(ascending=False)
    .head(30)
    .index.tolist()
)
vitals48_topitems
```
